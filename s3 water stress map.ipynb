```python
import pandas as pd
import plotly.express as px

file_path = "Aqueduct40_baseline_annual_y2023m07d05.csv"
df = pd.read_csv(file_path)

water = df[["name_0", "bws_score", "bws_cat", "bws_label"]].copy()
water.columns = ["country", "bws_score", "bws_cat", "bws_label"]

water = water.dropna(subset=["country"])

country_map = (
    water.groupby("country", as_index=False)
    .agg(
        water_stress_score=("bws_score", "max"),
        water_stress_cat=("bws_cat", "max")
    )
)

category_map = {
    0: "Low",
    1: "Low-medium",
    2: "Medium-high",
    3: "High",
    4: "Extremely high",
    5: "Arid / low water use"
}

country_map["water_stress_cat"] = country_map["water_stress_cat"].round().astype(int)
country_map["stress_category"] = country_map["water_stress_cat"].map(category_map)

category_order = [
    "Low",
    "Low-medium",
    "Medium-high",
    "High",
    "Extremely high",
    "Arid / low water use"
]

country_map["stress_category"] = pd.Categorical(
    country_map["stress_category"],
    categories=category_order,
    ordered=True
)

color_map = {
    "Low": "#f0f0f0",
    "Low-medium": "#cfc3f3",
    "Medium-high": "#5dcdfb",
    "High": "#4385f5",
    "Extremely high": "#313ad3",
    "Arid / low water use": "#d9d9d9"
}

fig = px.choropleth(
    country_map,
    locations="country",
    locationmode="country names",
    color="stress_category",
    hover_name="country",
    category_orders={"stress_category": category_order},
    color_discrete_map=color_map,
    custom_data=["stress_category", "water_stress_score"],
    title="<b>Global Water Stress</b>"
)

fig.update_traces(
    hovertemplate=
    "<b>%{hovertext}</b>"
    "<br><br>"
    "Water stress: %{customdata[0]}"
    "<br>"
    "Score: %{customdata[1]:.1f}"
    "<extra></extra>",
    
    marker_line_color="#fbf4ec",
    marker_line_width=0.4
)

fig.update_layout(
    hoverlabel=dict(
        bgcolor="#4c270d",
        bordercolor="#4c270d",
        font=dict(
            family="inter",
            size=12,
            color="#FFFFFF"
        ),
        align="left"
    )
)

fig.update_layout(
    width=800,
    height=650,
    margin=dict(l=20, r=20, t=120, b=60),
    paper_bgcolor="#fbf4ec",
    plot_bgcolor="white",
    legend_title_text="Baseline water stress",
   title=dict(
    text="Global Water Stress",
    x=0.02,
    xanchor="left",
    font=dict(
        family="Inter",
        size=48,
        color="#4c270d",
        weight="bold"
    )
),
    font=dict(
        family="inter",
        size=14,
        color="#4c270d"
    )
)

fig.update_geos(
    showframe=False,
    showcoastlines=False,
    showcountries=True,
    countrycolor="#181413",
    projection_type="natural earth",
    bgcolor="#fbf4ec",
    domain=dict(
        x=[1, 1],
        y=[1, 1]
    ),
    projection_scale=2,
    center=dict(lat=30, lon=30)
)
fig.add_annotation(
    text="Source: World Resources Institute (WRI), Aqueduct Water Risk Atlas 4.0",
    x=0,
    y=-0.12,
    xref="paper",
    yref="paper",
    showarrow=False,
    align="left",
    font=dict(
        family="inter",
        size=11,
        color="#6b625b"
    )
)

fig.show()
```

```python
fig.write_html("global_water_stress_map.html", include_plotlyjs="cdn")
```
